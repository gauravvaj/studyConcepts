## Matplotlib is a python library which is used for ploting various graphs .

```python
import matplotlib 
```

```python
import matplotlib.pyplot as plt # most of the plotting is insdide submodule called pyplot inside matplotlib.
import numpy as np
```

```python
xpoints = np.array([0, 6, 10,12])
ypoints = np.array([0, 250,20,24])

plt.plot(xpoints, ypoints)             # polt() functionis used to plot points on graph.
plt.show()
```

```python
ypoints = np.array([0, 250,20,24])
plt.plot( ypoints,marker='o')             # different markers.
plt.show()
plt.plot( ypoints,marker='*')
plt.show()
```

```python
plt.plot( ypoints,'*:r')             # * shows the marker,: shows dotted line and r shows red color.
plt.show()
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([12,14,16,20,22,23])

plt.plot(x, y)
plt.grid()                  # Making Grid 
plt.title("random Plot")    # Making title of plot
plt.xlabel("xpoints")     # Making labels for x-axis and y-axis.
plt.ylabel("ypoints")

plt.show()
```
